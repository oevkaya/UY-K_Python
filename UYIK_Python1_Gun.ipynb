DÖNGÜLER

Bir sürecin tekrar tekrar devam etmesi Python’da ‘döngü’ (loop) olarak adlandırılır. Python’da iki tane döngü bulunur: while ve for

```python
a = 1

while a < 10:
    a += 1
    print(a)
```

```python
a = 0

while a < 100:
    a += 1
    if a % 2 == 0:
        print(a)
```

for da tıpkı while gibi bir döngüdür. Yani tıpkı while döngüsünde olduğu gibi, programlarımızın birden fazla sayıda çalışmasını sağlar. Ancak for döngüsü while döngüsüne göre biraz daha yeteneklidir. while döngüsü ile yapamayacağınız veya yaparken çok zorlanacağınız şeyleri for döngüsü yardımıyla çok kolay bir şekilde halledebilirsiniz.

for döngülerinin söz dizimi şöyledir:

for değişken_adı in değişken:
    yapılacak_işlem

```python
liste = [1, 2, 3, 4, 5]

for x in liste:
    print(x)
```

```python
for x in range(0, 10):
    print(x)
```

```python
sayılar = "123456789"

for i in sayılar:
    if int(i) > 3:
        print(i)
```

LAMBDA FONKSİYONLARI

Lambda fonksiyonları Python programlama dilinin ileri düzey fonksiyonlarından biridir. 

Peki lambda fonksiyonları nedir ve ne işe yarar?

Lambda fonksiyonlarını, bir fonksiyonun işlevselliğine ihtiyaç duyduğumuz, ama konum olarak bir fonksiyon tanımlayamayacağımız veya fonksiyon tanımlamanın zor ya da meşakkatli olduğu durumlarda kullanabiliriz

```python
# Örneğin; 
x  = 10
x10 = lambda x: x + 10

# 'x' adlı bir parametre alan bir lambda fonksiyonu tanımla. Bu fonksiyon, bu
# 'x parametresine 10 sayısını eklesin.

x10(x)
```

```python
# def tanımı ile kıyaslama: 

def toplam(param1, param2):
  return param1 + param2

toplam(4,10)
```

```python
toplam = lambda param1, param2: param1 + param2

toplam(4,10)
```

RECURSIVE (ÖZYİNELEMELİ) FONKSİYONLAR

Python fonksiyonları, yukarıdaki örnekte de gördüğünüz gibi, nasıl başka fonksiyonları çağırabiliyorsa, aynı şekilde, istenirse, kendi kendilerini de çağırabilirler. İşte bu tür fonksiyonlara Python programlama dilinde ‘kendi kendilerini yineleyen’, veya daha teknik bir dille ifade etmek gerekirse ‘özyinelemeli’ (recursive) fonksiyonlar adı verilir.

```python
def azalt(s):
    if len(s) < 1:
        return s
    else:
        print(s)
        return azalt(s[1:])

print(azalt('Python'))

print(azalt('P'))
```
